```python
#Detectar si tiene valores NULL o Vacios
def detectar_null_vacio(df):
    valores_vacion_null = {}
    for column in df.columns:
        nulls = df[column].isnull().sum()
        empties = (df[column]== '').sum()
        total = nulls + empties
        if total > 0:
            valores_vacion_null[column] = total
            
    return valores_vacion_null
    

resultado = detectar_null_vacio(df)

if resultado:
    for campo,cantidad in  resultado.items():
        print(f'El campo {campo} tiene {cantidad} valores NULL o vacios')
else:
    print('No hay campos NULL o vacios ')
```

```python
for year in years:
    plt.figure(figsize=(12,6))
    
    # Supongamos que tienes un DataFrame llamado df con las columnas 'codigo_repartidor', 'facturas_entregadas', 'facturas_totales'
    df_filtered = df[df['YEAR'] == year]
    
    # Agrupar los datos por 'codigo_repartidor' y calcular la suma de 'facturas_entregadas' y 'facturas_totales'
    facturas_por_repartidor = df_filtered.groupby('ID_REPARTIDOR')[['CANT_FACT_ENTREGA_TOTAL', 'CANT_FACT_TOTALES', 'CANT_FACT_ENTREGA_PARCIAL','CANT_FACT_ENTREGA_MOTIVOS']].sum().reset_index()

    # Convertir los datos al formato largo para que seaborn pueda entenderlos mejor
    facturas_long = pd.melt(facturas_por_repartidor, id_vars=['ID_REPARTIDOR'], value_vars=['CANT_FACT_ENTREGA_TOTAL', 'CANT_FACT_TOTALES', 'CANT_FACT_ENTREGA_PARCIAL','CANT_FACT_ENTREGA_MOTIVOS'],
                            var_name='cantidad', value_name='tipo_factura')

    # Crear el gráfico de barras
    plt.figure(figsize=(12, 6))
    ax = sns.barplot(x='ID_REPARTIDOR', y='tipo_factura', hue='cantidad', data=facturas_long)

    # Personalizar el gráfico
    ax.set(xlabel='Código de Repartidor', ylabel='Cantidad', title=f'Cantidad de Facturas y Entregadas en el Año {year}')
    

    # Añadir etiquetas a las barras
    for container in ax.containers:
        ax.bar_label(container, size=10)

    # Ajustar el layout
    plt.tight_layout()
    plt.xticks(rotation=45)

    # Mostrar el gráfico
    plt.show()
```

```python
X_train = train_set.drop('MONTO_COBRADO',axis=1)
y_train = train_set['MONTO_COBRADO'].copy()

X_train.info()
```

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
lr_model.fit(X_train, y_train) # Construcción/inferencia del modelo
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

# Hago predecir el modelo generado sobre el conjunto de training
y_pred = lr_model.predict(X_train)

# 4. Evaluo la performance del modelo sobre el conjunto de training, mediante R2 y RMSE
r2 = r2_score(y_train, y_pred)
print("Train R2=", r2)
mse = mean_squared_error(y_train, y_pred)
print("Train RMSE=", np.sqrt(mse))
```

```python

# Tomo las variables regresoras en X (y descarto 'quality' de ese dataframe)
X_test = test_set.drop('MONTO_COBRADO', axis=1)
# X_test = X_test[['fixed acidity', 'residual sugar', 'alcohol']]
# Tomo la variable a dependiente/a predecir
y_test = test_set['MONTO_COBRADO'].copy()

# Hago predecir el modelo generado sobre el conjunto de test
y_pred = lr_model.predict(X_test)

# 4. Evaluo la performance del modelo sobre el conjunto de test, mediante R2 y RMSE
r2 = r2_score(y_test, y_pred)
print("Test R2=", r2)
mse = mean_squared_error(y_test, y_pred)
print("Test RMSE=", np.sqrt(mse))
```
